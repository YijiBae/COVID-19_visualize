```python
# 파일 읽기
import pandas as pd

new_df_2020 = pd.read_csv("위경도변환_2020.csv")

# 필요한 feature만 사용
new_df_2020  = new_df_2020 [['위도', '경도', 'total', 'etl_ymd']]
new_df_2020
```

```python
# 파일 읽기
import pandas as pd

new_df_2019 = pd.read_csv("위경도변환_2019.csv")

# 필요한 feature만 사용
new_df_2019  = new_df_2019 [['위도', '경도', 'total', 'etl_ymd']]
new_df_2019
```

# 전체 대상으로 2.5 변경

```python
grouped_2020 =new_df_2020.groupby(['위도','경도'])
grouped_2020_2=pd.DataFrame(grouped_2020.sum()['total']).reset_index()
grouped_2020_2['total2']=[float(x)/10000 for x in grouped_2020_2['total']]
```

```python
grouped_2019 =new_df_2019.groupby(['위도','경도'])
grouped_2019_2=pd.DataFrame(grouped_2019.sum()['total']).reset_index()
grouped_2019_2['total2']=[float(x)/10000 for x in grouped_2019_2['total']]
```

```python
new_grouped_2020 = grouped_2020_2[grouped_2020_2['total2'] >= 2.5]
new_grouped_2020 = new_grouped_2020.reset_index(drop = True)
```

```python
new_grouped_2019 = grouped_2019_2[grouped_2019_2['total2'] >= 2.5]
new_grouped_2019 = new_grouped_2019.reset_index(drop = True)
```

```python
new_grouped_2019
```

```python
new_grouped_2020
```

```python
# 2020 지도 그리기
from tqdm import tqdm_notebook
import folium

위도=[]
경도=[]
total_num=[]
for i in (range(len(new_grouped_2020))):
    total_num.append(new_grouped_2020['total2'][i])
    위도.append(new_grouped_2020['위도'][i])
    경도.append(new_grouped_2020['경도'][i])
```

```python
m=folium.Map(location=[36.797268,127.128546],zoom_start=17)
```

```python
for i in range(len(new_grouped_2020)):
    folium.CircleMarker([위도[i],경도[i]],radius=total_num[i],fill_color='#3186cc',color='#3186cc').add_to(m)    
```

```python
m.save('천안히트맵_2020_2.5.html')
```
