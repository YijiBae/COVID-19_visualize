```
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
fontpath = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
font = fm.FontProperties(fname=fontpath)
plt.rc('font', family='NanumBarunGothic')
mpl.font_manager._rebuild()

ratio = [36, 7, 1, 1, 8]
labels = ['contact with patient', 'overseas inflow', 'River of Grace Community Church', 'gym facility in Cheonan', 'etc']
colors = ['#0F498A', '#3C66AC', '#2291D0', '#1CBFDF', '#104a56']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 1}

plt.pie(ratio, autopct='%.0f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops, textprops=dict(color="w"))

plt.legend(labels ,
          title="감염 경로",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

#plt.show()
plt.savefig('/content/drive/MyDrive/Colab Notebooks/data science/result/감염경로.png')
```
