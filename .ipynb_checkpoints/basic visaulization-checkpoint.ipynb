파일 별도 저장

```python
df_2019 = pd.read_csv("data/위경도변환_2019.csv")
df_2020 = pd.read_csv("data/위경도변환_2020.csv")
```

```python
df_2019.head()
```

```python
#필요없는 값 삭제하기 
sum_df_2019  = df_2019 = df_2019.drop(['id', "x", "y",  "total", "admi_cd","etl_ymd"], axis = 1)
sum_df_2020 = df_2020.drop(['id', "x", "y", "timezn_cd", "total", "admi_cd","etl_ymd"], axis = 1)

# 연령별로 남/녀 값 합치기
sum_df_2019 = sum_df_2019.sum(axis = 0)
sum_df_2020 = sum_df_2020.sum(axis = 0)
```

```python
# 새로운 dataframe 생성
final_list = []
for i in range(0, 28):
    lists = []
    lists.append(sum_df_2019[i])
    lists.append(sum_df_2020[i])
    final_list.append(lists)
    
my_df = pd.DataFrame(data=final_list,columns=['2019', '2020'])
# 인덱스 재설정
my_df.index = ['남00','남10','남15', '남20', '남25','남30','남35','남40', '남45', '남50', '남55','남60','남65','남70', '여00','여10','여15','여20','여25','여30','여35','여40','여45','여50','여55','여60','여65','여70']
```

```python
# 전체 인구 비교
ratio = [42491, 38238, 37447, 27544, 25656, 22181]
labels = ['쌍용 2동', '부성 2동', '부성 1동', '쌍용 1동', '불당동', '쌍용 3동']
colors = ['#0F498A', '#3C66AC', '#2291D0', '#1CBFDF', '#104a56', '#cccccc']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 1}

plt.pie(ratio, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops, textprops=dict(color="w"))

plt.legend(labels ,
          title="동",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

#plt.show()
plt.savefig('result/동별 인구 수.png')
```

```python
# 평균 유동인구 비교 
group_df_2019=df_2019.groupby(df_2019['admi_cd']).sum()
group_df_2020=df_2020.groupby(df_2020['admi_cd']).sum()

sum_group_df_2019 = group_df_2019.drop(['id', "x", "y",  "total","etl_ymd", "timezn_cd", "위도", "경도"], axis = 1)
sum_group_df_2020 = group_df_2020.drop(['id', "x", "y",  "total","etl_ymd", "timezn_cd", "위도", "경도"], axis = 1)

final_result_2019 = sum_group_df_2019.mean(axis = 1).to_frame()
final_result_2020 = sum_group_df_2020.mean(axis = 1).to_frame()
```

```python
# 데이터 프레임화 
final_result_2019 = sum_group_df_2019.mean(axis = 1).to_frame()
final_result_2020 = sum_group_df_2020.mean(axis = 1).to_frame()

final_result_2019 = final_result_2019.reset_index()
final_result_2020 = final_result_2020.reset_index()

df_1 = pd.merge(final_result_2019, final_result_2020, right_index = True, left_index= True)

new_idx= ['쌍용1동', '쌍용2동', '쌍용3동', '불당동', '부성1동', '부성2동']
df_1['town'] = new_idx
```

```python
df_1 = df_1.set_index("town")
df_1 = df_1.drop(['admi_cd_x', 'admi_cd_y'], axis = 1)
df_1.columns = ['2019', '2020']
df_1
```

```python
# title = "전체 연령대별 유동 인구 수", 
mpl.rcParams.update({'text.color' : 'black'})

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
color = ['#2764AC', '#77A0CF']
df_1.plot(kind="bar", figsize=(25, 10), fontsize=20, color = color)
plt.legend(prop={'size': 20})
#plt.show()
plt.savefig('result/동별 유동인구.png')
```
